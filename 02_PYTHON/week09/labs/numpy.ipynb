> En NumPy, a las dimensiones se les conoce con el nombre de ejes( *axis** ) y al número de los ejes rango (**rank**).

- Array es un alias para referirse al tipo de objeto **numpy.ndarray**
Tenemos algunas de las *propiedades* de los arrays:
- *ndarray.ndim*: el número de ejes
- *ndarray.shape*: devuelve una tupla de enteros indicando longitud de las dimensiones de la matriz
- *ndarray.size*: total de elementos de una matriz

```python
# Ejemplo de creación de una matriz 3x2 (3 filas y 2 columnas) - serie unidimensional
matriz = np.arange(4*2)
print("Array unidimensional")
print(matriz)
```

```python
type(matriz)
```

> Para crear un array, tenemos diferentes opciones

```python
zeros = np.zeros(10)
zeros
```

```python
print(zeros.ndim) # dimensiones
print(zeros.size) # tamaño
print(zeros[0]) # primer elemento
print(type(zeros[0])) # tipo de valor
print(type(zeros)) # tipo de objeto
```

```python
# podemos cambiar cualquier de los valores accediendo por su posición
# Ejemplo queremos asignar 5 a la posición 4
zeros[4] = 5
zeros
```

```python
# asignar un valor 9 a la última posición
zeros[-1] = 9
zeros
```

```python
# transformamos los valores inicial de una matriz
matriz = np.arange(10,20) # start, end -1
matriz
```

```python
# Si queremos introducir solo valores impares desde 1 al 100
matriz_impar = np.arange(1, 101, 2) # start, end-1, step
matriz_impar
```

```python
# Podemos crear arrays desde listas de Python de varias dimensiones
lista_1 = [1, 2, 3]
lista_2 = [4, 5, 6]
matriz_listas = np.array([lista_1, lista_2])
matriz_listas
```

```python
# Podemos crear arrays desde listas de Python de varias dimensiones
lista_1 = [1, 2, 3]
lista_2 = [4, 5, 6]
matriz_listas = np.array(list(lista_1), list(lista_2)) # no admite la función list
matriz_listas
```

```python
# Que es lo mismo que crear
matriz_2 = np.array([[1,2,3],[4,5,6]]) # cada bloque de listas representan filas en una matriz
matriz_2
```

## Operaciones con matrices

```python
# construimos dos matrices
A = np.array([[1,0],[0,1]])
B = np.array([[1,2],[3,4]])
```

```python
A
```

```python
B
```

```python
# Suma de matrices
print(A+B)
print("*"*10)
# Resta de matrices
print(A-B)
print("*"*10)
# multiplicando matrices elemento por elemento
print(A*B)
print("*"*10)
# multiplicar matriz por matriz
print(A.dot(B))
print("*"*10)
# Potencia en matrices
print(A**2)
print("*"*10)
```

### Otra operaciones con Numpy, array y matrices

```python
# generamos una lista de 1mi
lst = list(range(1_000_000))
```

```python
lst[:20]
```

```python
%%timeit
for i in lst:
  doubles = []
  q = i**2
  doubles.append(q)
```

```python
%%timeit
# Para medir el tiempo en cada uno de los ciclos
list(map(lambda x: x*2, lst))
```

```python
%%timeit
doubles = [n*2 for n in lst]
```

```python
# Convertimos la lista generada en un array
arreglo_2 = np.array(lst)
arreglo_2
```

```python
%%time
doubles = arreglo_2 ** 2
doubles
```

```python
%%timeit
doubles = arreglo_2 ** 2
```
