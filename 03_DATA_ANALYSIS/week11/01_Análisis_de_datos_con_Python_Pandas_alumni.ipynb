Creamos un nuevo atributo `parsed_place` y transformamos CA California y NV Nevada
MX Mexico

````python
# Operamos con transformación de unos valores
df['parsed_place'] = df.place.str.replace(
    r'.* of ', '', regex=True # elimina sentencia of
    ).str.replace(
    r'CA$', 'California', regex=True
    ).str.replace(
        r'NV$', 'Nevada', regex=True
    ).str.replace(
        r'MX$', 'Mexico', regex=True
    ).str.replace(
        r' region$', '', regex=True
    ).str.replace(
        r'^.*, ', '', regex=True
    ).str.strip() # quita espacios
````

````python
df.parsed_place.sort_values().unique()
````

````python
# una válida alternativa es utilizar assign()
df.assign(
    in_ca=df.parsed_place.str.endswith('California'),
    in_alaska=df.parsed_place.str.endswith('Alaska')
).sample(5, random_state=17)
````

````python
# También podemos hacer uso de la función lambda()
df.assign(
    in_ca=df.parsed_place == 'California',
    in_alaska=df.parsed_place == 'Alaska',
    neither=lambda x: (~x.in_ca) & (~x.in_alaska)
).sample(5, random_state=17)
````

# Utilizar Apply - Transform

````python
import re

# Función para limpiar los nombres de place
def clean_place(row):
    # Buscamos cualquier caracter repetido del nombre
    if re.search(' of.*', row):
        # Extrae la posición inicial del patrón
        pos = re.search(' of.*', row).start()

        return row[:pos]
    else:
        return row
````

````python
df_raw['place_cleaned'] = df_raw['place'].apply(clean_place)
````
